```python
import pandas as pd

## Create a DataFrame from a dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}

df =  pd.DataFrame(data)
#Your code here

#Display the first 2 rows of the data frame
#Your code here
print("FIRST TWO ROWS:")
print(df.loc[[0,1]])
print()

#Print the age column
#Your code here
print("AGE_COLUMN:")
print(df["Age"])
print()

#Filter rows where age is greater than 26
#Your code here
print("ROWS WITH AGE GREATER THAN 26:")
# for i in range(3):
#   if df.iloc[i,1] > 26:
#     print(df.loc[i])
#     print()
print(df[df['Age'] > 26])
print()

#Add a new column 'Country' with the value 'India' for all rows
#Your code here
df["Country"] = ['India','India','India']
print(df)
print()



data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}

df2= pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age
#Your code here

x = df2['Age'].mean()
df2['Age'].fillna(x,inplace = True)
print(df2)
```

# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np #numpy module. Fill your code



xpoints=np.array([1,2,3,4])
ypoints=np.array([2,4,6,8])

#Plot these points without drawing a line
#Your code here
plt.figure(figsize=(10, 8))
plt.subplot(3, 2, 1)
plt.plot(xpoints, ypoints, 'o')
plt.title('Points without line')
plt.xlabel("xpoints")
plt.ylabel("ypoints")

#Plotting with marker: Plot these points with a marker(Star marker)
#Your code here
plt.subplot(3, 2, 2)
plt.plot(xpoints, ypoints, marker='*')
plt.title('Points with star marker')
plt.xlabel("xpoints")
plt.ylabel("ypoints")

#Using fmt format, add circular marker,red color and Dashed line
#Your code here
plt.subplot(3, 2, 3)
plt.plot(xpoints, ypoints, 'o--r')
plt.title('Red circular markers with dashed line')



#Add xlabel,ylabel and title for the plot.
#Your code here
plt.suptitle("Plot")
plt.xlabel("xpoints")
plt.ylabel("ypoints")

#Create a scatter plot for xpoints and ypoints
#Your code here
plt.subplot(3, 2, 4)
scatter = plt.scatter(xpoints, ypoints)
plt.title('Scatter plot')
plt.xlabel("xpoints")
plt.ylabel("ypoints")

 #Set color to the scatter plot. Blue,Green,Red and yellow color for each point respectively
colors = np.array(["Blue","Green","Red","Yellow"])
plt.subplot(3, 2, 6)
scatter = plt.scatter(xpoints, ypoints)
plt.scatter(xpoints, ypoints, c=colors)
plt.title('Scatter plot with colors')
plt.xlabel("xpoints")
plt.ylabel("ypoints")

plt.tight_layout()
plt.show()
```

# Miscellaneous Modules Implementation

```python
import random
import numpy as np

#Set the seed of random to 20
#Your code here
random.seed(20)

arr1 = np.array([1,24,31,45,73,81,94,25])

#Using the random module pick 4 different random numbers from arr1 and return their sum.
#Your code here
random_numbers = random.sample(list(arr1), 4)

# Calculate their sum
sum_of_random_numbers = sum(random_numbers)

print("Random numbers:", random_numbers)
print("Sum of random numbers:", sum_of_random_numbers)
```
